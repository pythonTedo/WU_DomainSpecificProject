```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Selenium WebDriver (you may need to adjust the path to your WebDriver)
driver = webdriver.Chrome()  # Or use 'webdriver.Firefox()' if you have Firefox WebDriver installed
driver.get("https://fashionunited.com/brands/all")

# Scroll to load all dynamic content
SCROLL_PAUSE_TIME = 2  # Adjust depending on the loading time of the site

last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # Scroll down to the bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for new content to load
    time.sleep(SCROLL_PAUSE_TIME)
    
    # Check if we've reached the bottom
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Parse the fully loaded page with BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")
driver.quit()  # Close the browser

base_url = "https://fashionunited.com"
brand_links = []
for li in soup.find_all("li"):
    a_tag = li.find("a")
    if a_tag and a_tag.get("href") and a_tag["href"].startswith("/companies/"):
        brand_links.append(base_url + a_tag["href"])

```

```python
# Function to extract brand details from a brand's page
def extract_brand_details(brand_url):
    response = requests.get(brand_url)
    brand_soup = BeautifulSoup(response.text, "html.parser")
    
    # Extract brand name
    brand_name = brand_soup.find("h1", text=True)
    brand_name = brand_name.text.strip() if brand_name else None
    
    # Extract type of products
    products = None
    products_section = brand_soup.find("dt", text="Products:")
    if products_section:
        products = products_section.find_next("dd").text.strip()
    
    # Extract description
    description = None
    description_divs = brand_soup.find_all("div", class_="css-14ed7fa e12cb4a14")

    if description_divs:
        # Combine text from all matching divs
        description = " ".join(div.text.strip() for div in description_divs)
    
    # Extract website
    website = None
    website_section = brand_soup.find("dt", text="Website:")
    if website_section:
        website = website_section.find_next("a").get("href")
    
    return {
        "Brand Name": brand_name,
        "Products": products,
        "Description": description,
        "Website": website
    }

# Extract details for each brand
brand_data = []
for link in brand_links:
    try:
        brand_details = extract_brand_details(link)
        brand_data.append(brand_details)
    except Exception as e:
        print(f"Error processing {link}: {e}")

# Save data to a pandas DataFrame
df = pd.DataFrame(brand_data)
```

## Data Gathering - Google Trends

```python
# import the TrendReq method from the pytrends request module
from pytrends.request import TrendReq
import time 
```

```python

# Initialize the pytrends API
pytrends = TrendReq(hl='en-US', tz=360)

def get_trend_data(args):
    list_failed = []
    # Keywords and settings#
    keywords = [args]
    time.sleep(30) 
    pytrends.build_payload(keywords, timeframe='today 12-m', geo='', gprop='')
    # Wait before fetching data
    # Add a delay to avoid being blocked
    
    try:
        region_data = pytrends.interest_by_region(resolution='COUNTRY')
        print(f"{args} passed")
        return region_data.to_dict()  
        
    except Exception as e:
        print(f"The brand {args} gave error!")
        print(f"Error: {e}")
        
    
```

```python
df1 = pd.read_excel("data\most_valuanble_brands.xlsx").drop("#", axis = 1)
df1
```

```python
import json

with open('data/data.json', 'r') as file:
    js = json.load(file)

brands = []
for i in js:
    brands.append(i.keys())

string_list = [list(dk)[0] for dk in brands]
```

```python
# Find missing values

missing_from_list = [value for value in df1['Name'].values if value not in string_list]
```

```python
# Batch processing with delay
batch_size = 5
results = []
for i in range(0, len(df1), batch_size):
    batch = df1.iloc[i:i+batch_size]
    for brand in batch['Name']:
        if brand not in string_list:
            trend_data = get_trend_data(brand)
            results.append(trend_data)
            time.sleep(10)  # Longer delay to avoid rate limits
        print(f"Brand {brand} already exists")
    print(f"Processed batch {i // batch_size + 1}")
```
